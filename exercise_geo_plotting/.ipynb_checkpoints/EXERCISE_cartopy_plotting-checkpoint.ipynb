```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./data/final_loc_data.csv")
pd.to_datetime(df.time)
df.head()
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection=ccrs.Mercator())

# Scatter data
# Data is in latlon so scatted command transform must be set to PlateCarree
# I don't know why longitude is off by 360 in this dataset but I think it is
ax.scatter(360-df['lon'], df['lat'], transform=ccrs.PlateCarree(), 
           c=df['temp'], s=df['precip'], cmap='magma')

ax.add_feature(cfeature.COASTLINE) # Add coastline
ax.gridlines()
plt.show()
```
